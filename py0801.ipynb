```
import re
text = "文本最重要的來源無疑是網格。我們要把網路中的文本獲取形成一個文本數據庫。利用一個爬蟲抓取到網路中的信息。爬蟲的策略有廣度爬取和深度爬取。根據用戶的需求，爬蟲可以有主題爬蟲和通用爬蟲之分。"
p_string = text.split("。")
regex = "^文本"
for line in p_string:
  if re.search(regex,line):
    print(line)
regex = "信息$"
for line in p_string:
  if re.search(regex,line):
    print(line)
```
